```python
import pandas as pd
import re
import numpy as np 
```

```python
file_path='tweet_result_19_.txt'
f=open(file_path,'r',encoding='latin1').read()
tw=re.split(r'\n+\n',f)
df=pd.DataFrame([lines.strip('\n') for lines in tw])
mask=(df[0].str.len()>1)

df=df.loc[mask]

df[0]=df[0].apply(lambda x :re.split('\n',x))
df['user']=df[0].apply(lambda x: x[0])
df['user_id']=df[0].apply(lambda x: x[1])
df['tag_num']=df[0].apply(lambda x: x[7] if len(x)>8 else 0)
df['tag_url']=df[0].apply(lambda x: x[8:]if len(x)>9 else 0)
df['tweet_content']=df[0].apply(lambda x: re.split('\s',x[6]) if len(x)>7 else None)
df_subset=df.dropna(subset=['user','user_id','tweet_content'])[['user','user_id','tweet_content','tag_num','tag_url']]
```
